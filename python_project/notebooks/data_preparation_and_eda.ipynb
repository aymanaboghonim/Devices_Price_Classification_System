# 1. Introduction


# Data Preparation and Exploratory Data Analysis for Device Pricing

Welcome to this comprehensive analysis where we aim to explore and prepare a dataset of mobile device specifications. This exploration is foundational for building a model that can predict the price range of these devices based on their characteristics.


# 2. Importing Necessary Libraries


First, let's import the necessary libraries that will help us in data manipulation, visualization, and statistical analysis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import joblib
```

# 3. Data Loading


We start by loading the training and testing datasets. Our goal here is to understand the basic structure of the data including column details and preliminary statistics.


```python
# Load the data
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')

# Display the first few rows of the training dataset
display(train_df.head())
```

# 4. Data Cleaning and Preprocessing


Effective data cleaning is crucial for accurate models. We will identify and fill missing values, remove any duplicates, and handle incorrect data formats.


```python
# Check for missing values in the training data
missing_values_train = train_df.isnull().sum()
print("Training data missing values in each column:\n", missing_values_train[missing_values_train > 0])

# Fill missing values with median for numerical columns in the training data
train_df.fillna(train_df.median(), inplace=True)

# Check for duplicates in the training data
duplicates_train = train_df.duplicated().sum()
print(f"Number of duplicate entries in training data: {duplicates_train}")

# Drop duplicates if any in the training data
train_df.drop_duplicates(inplace=True)

# Save the cleaned training data to a new CSV file
train_df.to_csv('../data/cleaned_train_data.csv', index=False)

# Check for missing values in the test data
missing_values_test = test_df.isnull().sum()
print("Test data missing values in each column:\n", missing_values_test[missing_values_test > 0])

# Fill missing values with median for numerical columns in the test data
test_df.fillna(test_df.median(), inplace=True)

# Check for duplicates in the test data
duplicates_test = test_df.duplicated().sum()
print(f"Number of duplicate entries in test data: {duplicates_test}")

# Drop duplicates if any in the test data
test_df.drop_duplicates(inplace=True)

# Save the cleaned test data to a new CSV file
test_df.to_csv('../data/cleaned_test_data.csv', index=False)
```

# 5. Feature Engineering and Normalization


## Normalize Numerical Features

To ensure all numerical features contribute equally to our model, we will standardize them. This avoids bias towards variables with higher magnitudes.


```python
# Identifying numerical columns
numeric_features = train_df.select_dtypes(include=['int64', 'float64']).columns.drop('price_range')

# Applying Standard Scaler
scaler = StandardScaler()
train_df[numeric_features] = scaler.fit_transform(train_df[numeric_features])
test_df[numeric_features] = scaler.transform(test_df[numeric_features])  # Use the same scaler to transform the test data

# Save the scaler
joblib.dump(scaler, '../models/scaler.pkl')

# Save the scaled data to a new CSV file
train_df.to_csv('../data/scaled_train_data.csv', index=False)
test_df.to_csv('../data/scaled_test_data.csv', index=False)
```

# 6. Exploratory Data Analysis (EDA)


## Exploratory Data Analysis (EDA)

In this section, we dive deep into the dataset through visualizations and data summaries to uncover patterns, detect outliers, and understand feature relationships.


### Summary Statistics


```python
# Displaying descriptive statistics
display(train_df.describe())
```

### Feature Distribution Visualization

To gain a clearer understanding of the distributions and to identify any potential skewness or outliers more distinctly, we'll use seaborn's `distplot`. This function will allow us to visualize both the histogram and the kernel density estimate (KDE) on the same plot.


```python
# Setting up the matplotlib figure
plt.figure(figsize=(16, 12))

# Create a FacetGrid to organize the plots
g = sns.FacetGrid(pd.melt(train_df[numeric_features]), col="variable", col_wrap=4, sharex=False, sharey=False)
g.map(sns.histplot, "value", kde=True, color="skyblue")

# Adding a title and adjusting spacing
g.fig.suptitle('Distribution of Numerical Features with Histogram and KDE', fontsize=16)
g.fig.subplots_adjust(top=0.95) # Adjust the top to make space for title

# Displaying the plots
plt.show()
```

### Correlation Analysis

Analyzing how features correlate with each other and especially with the target variable `price_range`.


```python
# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(train_df.corr(), annot=True, cmap='Blues', fmt=".2f")
plt.title('Feature Correlation Matrix')
plt.show()
```

### Pairwise Relationships

Investigating relationships between pairs of important features using pair plots.


```python
# Selecting a subset of features for pairwise plots
important_features = ['battery_power', 'ram', 'px_height', 'px_width', 'price_range']
sns.pairplot(train_df[important_features], hue='price_range', palette='viridis')
plt.suptitle('Pairwise Relationships Between Features', verticalalignment='top')
plt.show()
```
